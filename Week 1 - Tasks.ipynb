```python
from imutils import paths
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Task 1

Compute image descriptors (QS1)
â€¢ Color Histogram:
- gray level / concatenate color component histograms
- color space RGB, CieLab, YCbCr, HSV
- Compulsory to use 1D histograms!

```python
path = "../Data/qsd1_w1/"
```

```python
def imageRead(x = None):
    if x == None:
        imagePaths = list(paths.list_images('../Data/qsd1_w1'))
        return imagePaths
    else:
        image = cv2.imread(path + x)
        return image
        
```

```python
imagePaths = imageRead()
```

```python
Tot = len(imagePaths)
Cols = 5
Rows = Tot // Cols 

if Tot % Cols != 0:
    Rows += 1

Position = range(1,Tot + 1)
```

```python
index = 0
fig = plt.figure(1)
for images in imagePaths:
    image = cv2.imread(images)
    R, G, B = image[:,:,0], image[:,:,1], image[:,:,2]
    imgGray = (R + G + B) / 3
    imgGray = np.float32(imgGray)
    hist = cv2.calcHist([imgGray],[0],None,[256],[0,256])
    ax = fig.add_subplot(Rows,Cols,Position[index])
    ax.plot(hist)
    index += 1
```

```python
# different formula
index = 0
fig = plt.figure(1)
for images in imagePaths:
    image = cv2.imread(images)
    R, G, B = image[:,:,0], image[:,:,1], image[:,:,2]
    imgGray = 0.2126 * R + 0.7152 * G + 0.0722 * B
    imgGray = np.float32(imgGray)
    hist = cv2.calcHist([imgGray],[0],None,[256],[0,256])
    ax = fig.add_subplot(Rows,Cols,Position[index])
    ax.plot(hist)
    index += 1
```

```python
# different formula
index = 0
fig = plt.figure(1)
for images in imagePaths:
    image = cv2.imread(images)
    R, G, B = image[:,:,0], image[:,:,1], image[:,:,2]
    imgGray = 0.2126 * R + 0.7152 * G + 0.0722 * B
    imgGray = np.float32(imgGray)
    plt.subplot(Rows,Cols,Position[index])
    plt.imshow(imgGray, cmap = 'gray')
    plt.axis("off")
    index += 1

plt.show()
plt.close()
```
